### Load the dataset

```python
# Read a CSV file into a pandas DataFrame
world_df = pd.read_csv("gapminder.csv", index_col='country', sep=',')     #sep ',' nie potrwebny bo , jest standardowy. W CSV moze byc tez /t dla tab lub ;
world_df
```

### Expand the dataset

```python
# Calculate population in milions
world_df['pop_mil'] = world_df['population'] / 1000000
world_df.head(3)

# Map colors by continent
mapping = {
    'Asia': 'red',
    'Europe': 'green',
    'Africa': 'blue',
    'Americas': 'yellow',
    'Oceania': 'black'
}
world_df['color'] = world_df['cont'].replace(mapping)
world_df.head(3)
```

### Visualize the dataset

```python
# Create a basis scatterplot (bubble chart)
x_values = world_df['gdp_cap']
y_values = world_df['life_exp']
# plt.scatter(x_values, y_values)

#Size the bubbles
s_values = world_df['pop_mil']
# plt.scatter(x_values, y_values, s=s_values)

#Color the bubbles
c_values = world_df['color']
plt.scatter(x_values, y_values, s=s_values, c=c_values, alpha=0.5)

#Customize the plot
plt.xscale('log')
plt.xlabel('Income (USD)')
plt.ylabel('Life Expectancy (years)')
plt.title('2007 World Map')
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])
plt.text(2452, 64.7, 'India')
plt.text(4960, 72.9, 'China')
plt.grid(True)


#Display the plot
plt.show()
```
